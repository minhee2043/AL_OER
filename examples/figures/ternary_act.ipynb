```python
import matplotlib.colors as colors
from statistics import mean
from csv import writer
from csv import reader
import pickle
import math
import sys
import pickle
import os
import itertools as it
import numpy as np
from math import factorial
import pandas as pd
import matplotlib.pyplot as plt
import csv
import matplotlib.font_manager as fm
from scipy.optimize import minimize
# Font and style configuration
font_path = '/data2/minhee2043/NiCoFe/font/'
for font in fm.findSystemFonts(font_path):
    fm.fontManager.addfont(font)
    
config = {
    # Fonts.
    'font.family': 'Arial',
    'axes.linewidth': 0.75,
    # Figure element aesthetics.
    'lines.linewidth': 0.75,
    'lines.markersize': 3,
    'xtick.major.size': 2.5,
    'xtick.major.pad': 2.5,
    'ytick.major.size': 2.5,
    'ytick.major.pad': 2.5,    
    'figure.dpi': 300,
}

plt.rcParams.update(config)

kb=8.617333262*math.pow(10,-5)
T=300
E_opt=5.3

possibleratio=[]
for x in [x*0.05 for x in range(0,21)]:
    x=round(x,2)
    for y in np.arange(0,1.01-x,0.05):
        y=round(y,2)
        z=abs(1-x-y)
        z=round(z,2)
        possibleratio.append((x,y,z))

f=open('../batch15_act.txt','w')
activities=[]
for ratio in possibleratio:
    activity=0
    file=open(f'../batch15_count.csv')
    csvreader=csv.reader(file)
    for row in csvreader:
        possibility=math.pow(ratio[0],int(row[0]))*math.pow(ratio[1],int(row[1]))*math.pow(ratio[2],int(row[2]))
        boltz_up=-abs(float(float(row[3])-E_opt))
        boltz_down=kb*T
        final=possibility*math.exp(boltz_up/boltz_down)*int(row[-1])
        activity+=final
    f.write(str(ratio))
    f.write(str(activity))
    f.write('\n')
    activities.append(activity)

max_value=max(activities)
print('\n')
print('\n')
print(max_value)
print(min(activities))
f.write(str(max_value))
max_ind=activities.index(max_value)
print(possibleratio[max_ind])
f.write(str(possibleratio[max_ind]))
f.close()
possibleratio = np.array(possibleratio)
y0=0.5*np.sqrt(3)*possibleratio[:,2]
x0=possibleratio[:,1]+y0/np.sqrt(3)

plt.plot([0,1,0.5,0],[0,0,0.5*np.sqrt(3),0],'-k',zorder=1)
plt.annotate('Ni',(-0.08,-0.08),fontsize=12)
plt.annotate('Fe',(0.98,-0.08),fontsize=12)
plt.annotate('Co',(0.46,0.5*np.sqrt(3)+0.03),fontsize=12)
plt.scatter(x0,y0,s=16,zorder=2,c=activities,cmap='viridis')
plt.colorbar(label='Activity')
plt.xticks([])
plt.yticks([])
plt.xlim(-0.1,1.1)
plt.ylim(-0.1,0.5*np.sqrt(3)+0.1)
plt.show()
plt.savefig('batch15_act.png', transparent=False, bbox_inches='tight')
```
