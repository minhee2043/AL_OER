## Example 1: Extract Features from DFT Structure

````python
from active_learning import Slab
import ase.io

# Load structure
atoms = ase.io.read('your_structure.traj')

# Extract features
slab = Slab(atoms)
features = slab.features(['Ni', 'Fe', 'Co'])

print(f"Features: {features}")
````

## Example 2: Complete Workflow

````python
from active_learning import (
    DataspaceGenerator,
    ActiveLearner,
    ActivityCalculator
)

# Step 1: Generate dataspace
print("Step 1: Generating dataspace...")
gen = DataspaceGenerator()
gen.generate('GPRdataspace.csv')

# Step 2: Train models
print("\nStep 2: Training models...")
learner = ActiveLearner()
learner.load_data('DFT_O_all.csv', 'DFT_OH_all.csv')
learner.train()

# Step 3: Get suggestions
print("\nStep 3: Getting suggestions...")
suggestions = learner.suggest_next_batch(30)

# Step 4: Calculate activity
print("\nStep 4: Calculating activity...")
calc = ActivityCalculator()
calc.calculate_and_plot('batch15_count.csv', 'activity.png')

print("\n✓ Done!")
````
